```python
name = 'Tweets_data/Annoted/sub_tweet_03_annoted.csv'
test = pd.read_csv(name, sep='\t')
test = test.drop(test.columns[6], axis=1)
test = test.drop(test.columns[0], axis=1)
test
```

```python
import math
track = pd.read_csv('Model/trackingreboot_V1.csv', header=None, sep=',')
step = track[track.columns[1]].tolist()
train_loss = track[track.columns[2]].tolist()
tmp_test = track[track.columns[3]].tolist()

# Adapt values
test_loss_tmp = []
for itm in tmp_test:
    tmp = itm.replace('tensor(', '')
    test_loss_tmp.append(float(tmp))
test_loss = []
test_loss_idx = []
for i in range(0, math.floor(len(test_loss_tmp)/3)):
    test_loss.append(np.mean(test_loss_tmp[i*3:(i+1)*3]))
    test_loss_idx.append(np.mean(step[i*3:(i+1)*3]))
# plot it 
plt.plot(step, train_loss, c='Red', label='Train Loss')
plt.plot(test_loss_idx, test_loss, c='Green', label='Test Loss')
plt.legend()
plt.title('CamemBERT Fine-Tuning')
plt.xlabel('Batch index (batch-size = 20)')
plt.ylabel('Loss')
#plt.show()
plt.savefig('Figs/Learning_track.png')
plt.close()
```
